```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

# lmplot

```python
sns.lmplot(x='total_bill', y='tip', data=tips, markers='s')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', 
           markers=['o','v'], scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, 
           col='time', row='day')
```

### Value count based on "time" and "day"

```python
lunch = tips[tips['time']=='Lunch']
dinner = tips[tips['time']=='Dinner']
```

```python
lunch_count = {}
dinner_count = {}

for day in lunch['day']:
    lunch_count[day] = lunch_count.get(day, 0) + 1
for day in dinner['day']:
    dinner_count[day] = dinner_count.get(day, 0) + 1
```

```python
lunch_count
```

```python
dinner_count
```
