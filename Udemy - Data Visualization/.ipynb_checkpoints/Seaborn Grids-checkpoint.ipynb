```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# PairGrid

### PairPlot

```python
sns.pairplot(iris)
```

### Customize

```python
grid = sns.PairGrid(iris)
# grid.map(plt.scatter)
grid.map_diag(sns.distplot)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
```

# FacetGrid

```python
grid_2 = sns.FacetGrid(tips, row='sex', col='time')
# the number of rows must be equal to columns
grid_2.map(sns.distplot, 'total_bill')
```

```python
grid_2 = sns.FacetGrid(tips, row='sex', col='time')
# the number of rows must be equal to columns
grid_2.map(plt.scatter, 'total_bill', 'tip')
```
