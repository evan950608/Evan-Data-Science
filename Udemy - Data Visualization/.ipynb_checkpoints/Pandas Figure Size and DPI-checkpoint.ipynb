```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# fig = plt.figure(figsize=(8,2))
# # figsize is calculated in inches
# axes = fig.add_axes([0, 0, 1, 1])
# axes.plot(x, y)

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4))
axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig_2 = plt.figure()
axes_2 = fig_2.add_axes([0, 0, 1, 1])

axes_2.plot(x, x**2, label='f(x) = x**2')
axes_2.plot(x, x**3, label='f(x) = x**3')

axes_2.legend(loc=0)    # loc=0, automatically chooses best location
# axes_2.legend(loc=(0.2,0.2))
```
