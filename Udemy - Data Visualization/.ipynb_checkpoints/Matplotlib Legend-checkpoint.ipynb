# Legends

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 11)
```

```python
fig_2 = plt.figure()
axes_2 = fig_2.add_axes([0, 0, 1, 1])

axes_2.plot(x, x**2, label='f(x) = x**2')
axes_2.plot(x, x**3, label='f(x) = x**3')

axes_2.legend(loc=0)    # loc=0, automatically chooses best location
# axes_2.legend(loc=(0.2,0.2))
```

```python
fig_2 = plt.figure()
axes_2 = fig_2.add_axes([0, 0, 1, 1])

axes_2.plot(x, x**2, label='f(x) = x**2')
axes_2.plot(x, x**3, label='f(x) = x**3')

axes_2.legend(loc='best')    # loc=0, automatically chooses best location
# axes_2.legend(loc=(0.2,0.2))
```

# Location

```python
# ‘best’            0
# ‘upper right’	    1
# ‘upper left’	    2
# ‘lower left’	    3
# ‘lower right’	    4
# ‘right’	        5
# ‘center left’	    6
# ‘center right’	7
# ‘lower center’	8
# ‘upper center’	9
# ‘center’	       10
```
