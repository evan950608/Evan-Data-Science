```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

# Figure Size & DPI

```python
fig = plt.figure(figsize=(5,5))
# width and height of the figure in inches
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('y = x**2')
```

```python
fig_3, axes_3 = plt.subplots(nrows=2, ncols=1, figsize=(8,4))
axes_3[0].plot(x, y)
axes_3[1].plot(y, x)

plt.tight_layout()
```
