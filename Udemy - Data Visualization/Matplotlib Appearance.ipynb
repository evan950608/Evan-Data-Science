```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
```

# Color

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, color='#FF8C00')    # RGB Hex color code
axes.plot(x, x+1, color='blue')
# 'red', 'blue', 'green', 'yellow', 'brown', 'orange', 'purple', 'silver', 'gold', 'violet', 'pink', 'gray', 'black', 'white'
```

# Linewidth, Transparency

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x+1, color='b', lw=1)
axes.plot(x, x+2, color='b', lw=2)
axes.plot(x, x+3, color='b', lw=2, alpha = 0.5)
axes.plot(x, x+4, color='b', lw=3)
# Transparency: alpha, range[0, 1]
# linewidth = lw
```

# Linestyle

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x, color='purple', lw=2, ls='-')
axes.plot(x, x+1, color='purple', lw=2, ls='--')
axes.plot(x, x+2, color='purple', lw=2, ls='-.')
axes.plot(x, x+3, color='purple', lw=2, ls=':')
axes.plot(x, x+4, color='purple', lw=2, ls='steps')

# linestyle = ls
```

# Marker

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x, color='brown', lw=1, marker='o', markersize=5)
axes.plot(x, x+1, color='brown', lw=1, marker='+', markersize=5)
axes.plot(x, x+2, color='brown', lw=1, marker='*', markersize=5)
axes.plot(x, x+3, color='brown', lw=1, marker=1, markersize=5)
axes.plot(x, x+4, color='blue', lw=2, marker='o', markersize=10, markerfacecolor='yellow')
axes.plot(x, x+5, color='blue', lw=2, marker='o', markersize=10, 
          markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='red')
```
