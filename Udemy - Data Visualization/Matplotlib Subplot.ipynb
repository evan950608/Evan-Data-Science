```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

# Subplot

```python
fig, axes = plt.subplots()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()    # deals with overlapping problems
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First')
axes[1].plot(y, x)
axes[1].set_title('Second')

plt.tight_layout()
```

```python
axes
```

```python
fig_2, axes_2 = plt.subplots(nrows=1, ncols=2)
for ax in axes_2:
    ax.plot(x, y)
    ax.set_title('y = x**2')
```
