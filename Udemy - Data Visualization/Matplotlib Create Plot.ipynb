```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

# Functional Method

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('y = x**2')
```

```python
# subplot(number of rows, number of columns, index)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')    # color is 'red'

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')    # color is 'blue'
```

# Object Oriented Method

```python
fig = plt.figure()
# create a figure object
# like a blank canvas, ready to add things on it

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# [left, bottom, width, height]
# each value between 0 and 1

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('y = x**2')
```

```python
fig_2 = plt.figure()
axes_1 = fig_2.add_axes([0.1, 0.1, 0.8, 0.8])
axes_2 = fig_2.add_axes([0.2, 0.5, 0.4, 0.3])
    # 20% from the left in
    # 50% from the bottom up
    # takes up 40% of canvas width
    # takes up 30% of canvas height
axes_3 = fig_2.add_axes([0.5, 0.1, 0.4, 0.3])

axx = (axes_1, axes_2, axes_3)
for ax in axx:
    ax.plot(x, y)
    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_title('y = x**2')
```
