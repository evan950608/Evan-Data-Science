```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
# Seaborn comes with built-in data sets!
```

```python
tips.head()
```

# DistPlot

```python
sns.distplot(tips['total_bill'], bins=20)
# bins越大，級距切的越細
```

```python
sns.distplot(tips['tip'], kde=False)
```

# JointPlot

### kind = scatter

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
# x=column_name, y=column_name, data=dataframe_name
```

### kind = hex

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
# x=column_name, y=column_name, data=dataframe_name
```

### kind = reg

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
# draw a regression line
```

### kind = kde

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
# KDE: Kernel Density Estimation
```

# PairPlot

```python
# lists every possible combination
# quickly visualize data
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='time')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

# RugPlot

```python
sns.rugplot(tips['total_bill'])
# draw a dash mark for every point on a univariate distribution
```

# KDE-Plot

```python
sns.kdeplot(tips['total_bill'])
```

# KDE + Rug

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
