```python
import pandas as pd
```

```python
ecom = pd.read_csv(r"C:\PyCodes\Sample - PythonDataScience&MachineLearning\Python-for-Data-Analysis\Pandas\Pandas Exercises\Ecommerce Purchases.csv")
```

```python
list(ecom.columns)
```

```python
ecom['Browser'] = ecom['Browser Info'].apply(lambda x: x.split('/')[0])
```

```python
prices = list(ecom['Purchase Price'])
```

```python
interval = []
for price in prices:
    for n in range(10):
        if 10*n <= price and price < 10*(n+1):
            interval.append('{0} ~ {1}'.format(10*n, 10*(n+1)))
            break
```

```python
ecom['Interval'] = interval
```

```python
ecom.head()
```

```python
ecom.groupby(['Browser', 'Interval']).count()['Address']
```

```python
temp = pd.DataFrame(ecom.groupby(['Browser', 'Interval']).count()['Address'])
```
